# Intro to pyplot
1. matplotlib.pyplot is a collection of functions.
2. Each pyplot function makes some change to a figure.
     * creates a figure.
     * creates a plotting area in a figure. 
     * plots some lines in a plotting area. 
     * decorates the plot with labels, etc.

### Types of inputs to plotting functions.
* Plotting functions expect numpy.array or numpy.ma.masked_array as input, or objects that can be passed to numpy.asarray.
* Classes that are similar to arrays ('array-like') such as pandas data objects and numpy.matrix may not work as intended.
* Common convention is to convert these to numpy.array objects prior to plotting.

## Example - We will plot the line chart for developers and their salaries by their ages.

```python
dev_Age = [24,25,26,27,28,29,30]
dev_Salary = [30000,40000,50000,70000,80000,100000,120000]
```

```python
# now we will plot age on x-axis and salary on y-axis.
plt.plot(dev_Age, dev_Salary)   # If i run this now it will not show any plot.
plt.show()  # this will create my plot
```

This is my basic plot. On my x-axis ages are there and on y-axis salary but the meaning is hidden without labels. To add labels on both axis we can use **xlabel** and **ylabel** method.

```python
plt.plot(dev_Age, dev_Salary)
plt.xlabel('Ages')
plt.ylabel('Average Salary') 
plt.show()
```

Now this makes more sense. A title would add more sense to the plot. To add title to your plot you can use **title** method.

```python
plt.plot(dev_Age, dev_Salary)
plt.xlabel('Ages')
plt.ylabel('Average Salary')
plt.title('Salary in Rs by Age')
plt.show()
```

Now this looks more informative.

We can compare other domain salaries by same age. So we will take another example.

```python
SE_Salary = [25000,35000,40000,50000,70000,85000,100000]
```

```python
plt.plot(dev_Age, dev_Salary)
plt.plot(dev_Age, SE_Salary)
plt.xlabel('Ages')
plt.ylabel('Average Salary')
plt.title('Salary in Rs by Age')
plt.show()
```

As we can see two lines for the salary o two domains but we could not figure out which one belongs tow which one.
To indicate which line belongs to which domain, we can use **legend** method of plotly class.

I could add labels for legend as a list of values and order in the order they were added to the plot so developer wwere added first then software engineers.

```python
plt.plot(dev_Age, dev_Salary) 
plt.plot(dev_Age, SE_Salary)
plt.xlabel('Ages')
plt.ylabel('Average Salary')
plt.title('Salary in Rs by Age')
plt.legend(['Developers','Software Engineers']) # You need to enter arguments in order in which they were defined.
plt.savefig('plt1.png')
plt.show()
```

Now plot is more clear.

Other way to add legends to plot is we can pass label argument to plot method. This method is less error prone.

```python
plt.plot(dev_Age, dev_Salary, label='Developers')
plt.plot(dev_Age, SE_Salary, label='Software Engineers')
plt.xlabel('Ages')
plt.ylabel('Average Salary')
plt.title('Salary in Rs by Age')
# plt.legend(['Developers','Software Engineers'])
plt.legend()   # I will leave it without argument
plt.show()
```

We can change the formatting and style as well of this plot. Like we can chnage the color of lines and other stuffs.

We can do this by passing all of the information at once in a format string after our Y values in our plot method.

For more reference : https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

# Format Strings

A format string consists of a part for color, marker and line:
   *   format = '[marker][line][color]'

```python
plt.plot(dev_Age, dev_Salary, 's--b', label='Developers')
plt.plot(dev_Age, SE_Salary, 'o--g', label='Software Engineers')
plt.xlabel('Ages')
plt.ylabel('Average Salary')
plt.title('Salary in Rs by Age')
# plt.legend(['Developers','Software Engineers'])
plt.legend()   # I will leave it without argument
plt.show()
```

I can pass the format info in other way as well

```python
plt.plot(dev_Age, dev_Salary, color= 'b', linestyle='--', marker='s', label='Developers')
plt.plot(dev_Age, SE_Salary, color= 'g', linestyle='--', marker='o', label='Software Engineers')
plt.xlabel('Ages')
plt.ylabel('Average Salary')
plt.title('Salary in Rs by Age')
# plt.legend(['Developers','Software Engineers'])
plt.legend()   # I will leave it without argument
plt.show()
```
